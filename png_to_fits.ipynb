**PNG to FITS**

```python
import os
import sys
import numpy as np
import astropy.io.fits as pyfits
import imageio
from time import gmtime, strftime
```

Se lee el archivo png y se separa su nombre y extensión

```python
file = 'mock_36424930442626155.png'
inbase_name, inbase_ext = os.path.splitext(os.path.basename(file))
outfile = inbase_name + '.fits'
```

```python
inimage = imageio.imread(file, mode = 'F')
outimage = np.flipud(inimage)
```

Se crea un archivo FITS

```python
outlist = pyfits.PrimaryHDU(outimage.astype('float32'))
```

Se guarda la el tiempo se creación del archivo.

```python
file_time = strftime('%Y-%m-%d %H:%M:%S', gmtime())
```

Se guarda la información del archivo png en la cabecera del archivo FITS

```python
outhdr = outlist.header
outhdr['DATE'] = file_time
outhdr['history'] = 'Generated by fits'
outhdr['history'] = 'Image_file_' + file
```

Se agrega una nueva carta al header.

```python
# Lens parameters
c1 = pyfits.Card('theta_E', 1, 'Einstein Radius')
c2 = pyfits.Card('e1', 0.1, 'elpticity1')
c3 = pyfits.Card('e1', 0.1, 'elpticity2')

# Shear components
c4 = pyfits.Card('gamma1', 0.05, 'first shear component')
c5 = pyfits.Card('gamma2', 0.05, 'second shear component')
parameters = [c1, c2, c3, c4, c5]

for parameter in parameters:
    outhdr.append(parameter, end = True)
```

Se guarda el archivo FITS

```python
outlist.writeto(outfile, overwrite = True)
```
